```python
import pandas as pd
import numpy as np
import umap
import hdbscan
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

**Первый курс PCA**

```python
data = pd.read_csv("first_courses.csv")
data
```

```python
pca = PCA(n_components=2)
x = pca.fit_transform(data.iloc[:, 1:])
x = pd.DataFrame(x)
x.columns = ["v1", "v2"]
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=4).fit(x)
labels = dbscan.labels_
data["cluster"] = dbscan.labels_
data
```

```python
num_c = set(labels)
num_c
```

```python
clusterColor = {-1:"magenta"}
colors = [clusterColor[label] for label in labels]
plt.figure(figsize=(5, 5))
plt.scatter(x["v1"], x["v2"], c=colors)
plt.title("DBSCAN on all first course")
plt.show()
```

```python
clusterColor = {-1:"magenta"}
colors = [clusterColor[label] for label in labels]
plt.figure(figsize=(8, 8))
subject = data["Предмет"]
fig, ax = plt.subplots()
z = x["v1"]
y = x["v2"]
ax.scatter(z, y, c=colors)
for i, txt in enumerate(subject):
    ax.annotate(txt, (z[i], y[i]))
plt.title("DBSCAN на первом курсе")
plt.show()
```
